```python
# Home Appliances Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import Dataset
df = pd.read_csv('../data/home_data.csv')
```

```python
df_people_stay = df.groupby(['no_of_people', 'time_stayed_mins']).sum()['power']
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
df_people_stay.plot(kind='line')
plt.show()
```

```python
df_morning = df[df['from_time'] == '00:00:00' ]
df_people_room = df_morning.groupby(['room', 'time_stayed_mins']).sum()['no_of_people']
```
