```python
# Home Appliances Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
# Import Dataset
df = pd.read_csv('../data/home_data.csv')
```

```python
df_devices = df.groupby(['from_time', 'device', 'type']).mean()['power']
```

```python
from_times = df['from_time'].unique()
devices = df['device'].unique()
types = df['type'].unique()
```

```python
dd = {}
types_dict = {}
devices_dict = {}
from_times_dict = {} 

for i in types:
    types_dict[i] = 0
 
for i in devices:
    devices_dict[i] = types_dict
    
for i in from_times:
    from_times_dict[i] = devices_dict

dd = from_times_dict
# dd
```

```python
for index, row in df.iterrows():
#     print(row['from_time'], row['device'] ,row['type'], row['power'])
    dd[row['from_time']][row['device']][row['type']] = row['power']
dd
```
