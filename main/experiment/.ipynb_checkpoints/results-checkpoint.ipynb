```python
# Home Appliances Data
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
# Import Dataset
df_home_data_test = pd.read_csv('../../data/home_data_predict.csv')
df_home_priority = pd.read_csv('../../data/home_priority.csv')
date_time_group = pd.read_csv('../../data/date_time_group.csv')
final_output_group = pd.read_csv('../../data/final_output_group.csv')
```

```python
date_time_group.groupby('date').sum()['power']
powers = np.array(date_time_group.groupby('date').sum()['power'])
dates = np.array(date_time_group.groupby('date').sum()['power'].index)
dates = [datetime.datetime.strptime(date, '%Y-%m-%d').strftime('%d-%m-%Y') for date in dates]
```

```python
final_output_group.groupby('date').sum()['power']
final_powers = np.array(final_output_group.groupby('date').sum()['power'])
final_dates = np.array(final_output_group.groupby('date').sum()['power'].index)
final_dates = [datetime.datetime.strptime(date, '%d-%m-%Y').strftime('%d-%m-%Y') for date in final_dates]
```

```python
total_power_consumption = date_time_group['power'].sum()
print("Total: ", total_power_consumption)
```

```python
final_total_power_consumption = final_output_group['power'].sum()
print("Total: ", final_total_power_consumption)
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(np.arange(len(dates)), dates, rotation=30, fontsize=8)
plt.title('Power Consumption across a month', fontsize=20)
plt.xlabel('Dates', fontsize=20)
plt.ylabel('Power Consumption (W)', fontsize=20)
plt.plot(dates, powers, marker='o', color='r', label='Original')
plt.plot(final_dates, final_powers, marker='o', color='b', label='Optimized')
plt.legend(loc='best')

plt.show()
```
